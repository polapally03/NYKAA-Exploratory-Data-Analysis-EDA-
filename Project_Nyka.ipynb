```python
Product_Name = []
Price = []
SellingPrice=[]
Discount=[]
RatingCount = []
for i in range(1, 50):
    print('-------------------------------------------------------')
    start = time.time()
    URL = "https://www.nykaa.com/bath-body/body-care/body-lotions-moisturizers/c/371?search_redirection=True".format(i)
    page = requests.get(URL)
    page_content = page.text
    soup = BeautifulSoup(page_content)
    #print(soup.text)
    for x in soup.find_all('div', attrs ={'class':"css-ifdzs8"}):
        PName = x.find('div', attrs = {'class': "css-xrzmfa"})
        Price1 = x.find('span', attrs = {'class': 'css-17x46n5'})
        Selling1 = x.find('span', attrs = {'class': 'css-111z9ua'})
        Discount1 = x.find('span', attrs = {'class': 'css-cjd9an'})
        Rating1 = x.find('span', attrs = {'class': "css-1qbvrhp"})
        Volume1 = x.find('span', attrs = {'class': "css-1ctpgu6"})
        if PName is None:
            Product_Name.append(np.nan)
        else:
            Product_Name.append(PName.text)
            
        if Price1 is None:
            Price.append(np.nan)
        else:
            Price.append(Price1.text)
            
        if Selling1 is None:
            SellingPrice.append(np.nan)
        else:
            SellingPrice.append(Selling1.text)
            
        if Discount1 is None:
            Discount.append(np.nan)
        else:
            Discount.append(Discount1.text)
            
        if Rating1 is None:
            RatingCount.append(np.nan)
        else:
            RatingCount.append(Rating1.text)
                
        
    end = time.time()
    print("Page {} is completed in {} seconds".format(i, end-start))
```

```python
data = {'ProductName':Product_Name, 'Price':Price, 'SellingPrice':SellingPrice, 'Discount':Discount,'RatingCount':RatingCount}
df= pd.DataFrame(data)
```

```python
df
```

```python
df['Brand'] = df['ProductName'].str.split().str[0]
```

```python
df
```

```python
df["Price"] = (df["Price"].str.replace(r'[^0-9.]', '', regex=True))
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
df
```

```python
df["SellingPrice"] = (df["SellingPrice"].astype(str).str.replace(r'[^0-9.]', '', regex=True))
df["SellingPrice"] = pd.to_numeric(df["SellingPrice"], errors="coerce")

df["Discount"] = (df["Discount"].astype(str).str.replace(r'[^0-9.]', '', regex=True))
df["Discount"] = pd.to_numeric(df["Discount"], errors="coerce")
```

```python
df
```

```python
df["Savings"] = df["Price"] - df["SellingPrice"]
df["PriceCategory"] = pd.cut(df["Price"],bins=[0, 500, 1000,1200],labels=["Budget", "Mid-range", "Premium"])
```

```python
df
```

```python
new_order = ["ProductName",
    "Brand",
    "Price",
    "SellingPrice",
    "Discount",
    "Savings",
    "PriceCategory",
    "RatingCount"]
df = df[new_order]
```
