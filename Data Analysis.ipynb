```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("BodyLotion.csv")
df
```

```python
#Univariate(Histogram & KDE)
#Histogram & Kde for Selling Price
plt.figure(figsize=(5,3))
sns.histplot(df['SellingPrice'],bins=30,kde=True)
plt.title("Distribution of Selling Price")
plt.show()

#Histogram & Kde for Price
plt.figure(figsize=(5,3))
sns.histplot(df['Price'],bins=30,kde=True)
plt.title("Distribution of Price")
plt.show()

#Histogram & Kde for RatingCount
plt.figure(figsize=(5,3))
sns.histplot(df['RatingCount'],bins=30,kde=True)
plt.title("Distribution of RatingCount")
plt.show()

plt.figure(figsize=(5,3))
sns.histplot(df['Discount'],bins=30,kde=True)
plt.title("Distribution of Discount")
plt.show()

plt.figure(figsize=(5,3))
sns.histplot(df['Savings'],bins=30,kde=True)
plt.title("Distribution of Savings")
plt.show()
```

```python
#Univaiate (Box plot)
#Boxplot for Discount Price
plt.figure(figsize=(5,3))
sns.boxplot(x=df['Discount'])
plt.title("BoxPlot of Discount")
plt.show()

#Boxplot for  selling Price
plt.figure(figsize=(5,3))
sns.boxplot(x=df['SellingPrice'])
plt.title("BoxPlot of SellingPrice")
plt.show()

plt.figure(figsize=(5,3))
sns.boxplot(x=df['RatingCount'])
plt.title("BoxPlot of RatingCount")
plt.show()

plt.figure(figsize=(5,3))
sns.boxplot(x=df['Savings'])
plt.title("BoxPlot of Savings")
plt.show()

plt.figure(figsize=(5,3))
sns.boxplot(x=df['Price'])
plt.title("BoxPlot of Price")
plt.show()
```

```python
# Countplot for PriceCategory
plt.figure(figsize=(5,3))
sns.countplot(data=df, x="PriceCategory", order=df['PriceCategory'].value_counts().index)
plt.title("Count of Products by PriceCategory")
plt.show()
```

```python
# Pie chart for Brands
brand_counts = df['Brand'].value_counts()
plt.figure(figsize=(7,5))
plt.pie(brand_counts, labels=brand_counts.index, autopct="%2.2f%%", startangle=45)
plt.title("Brands Distribution")
plt.show()
```

```python
# Bivariate Analysis
# ------------------

# Selling Price vs Discount Price
plt.figure(figsize=(5,4))
sns.scatterplot(data=df, x="SellingPrice", y="Discount", hue="PriceCategory")
plt.title("Selling Price vs Discount Price")
plt.show()

# Savings by PriceCategory
plt.figure(figsize=(5,4))
sns.boxplot(data=df, x="PriceCategory", y="Savings")
plt.title("Savings by PriceCategory")
plt.show()

# Ratings Count vs PriceCategory
plt.figure(figsize=(5,4))
sns.barplot(data=df, x="PriceCategory", y="RatingCount", estimator='mean')
plt.title("Average Rating Count by PriceCategory")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Brand", hue="PriceCategory")
plt.title("Brand by PriceCategory Count")
plt.show()
```

```python
# Multivariate Analysis
# =========================

# Correlation Heatmap for numerical columns
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Crosstab: Brand vs PriceCategory
brand_price = pd.crosstab(df['Brand'], df['PriceCategory'])
print("\nBrand vs PriceCategory Crosstab:\n", brand_price)

# Pivot Table: Average Discount by Brand & PriceCategory
pivot_table = df.pivot_table(values="Savings", index="Brand", columns="PriceCategory", aggfunc="mean")
print("\nPivot Table - Avg Savings by Brand & PriceCategory:\n", pivot_table)

# Groupby: Discount  & Ratings count per Brand
groupby_discprice_ratings = df.groupby("Brand")[["Discount", "RatingCount"]].mean().reset_index()
print("\nGroupby - Discount & Rating\n",groupby_discprice_ratings)
```

```python
# Groupby Example
brand_summary = df.groupby("Brand").agg({
    "SellingPrice": "mean",
    "Discount": "mean",
    "RatingCount": "sum"
}).sort_values("SellingPrice", ascending=False)

print(brand_summary.head(10))

# Pivot Table Example (PriceCategory vs Average Ratings)
pivot = pd.pivot_table(df, 
                       values="RatingCount", 
                       index="PriceCategory", 
                       columns="Brand", 
                       aggfunc="mean")
print(pivot)

# Visualization: Boxplot (Selling Price by Price Band)
plt.figure(figsize=(5,3))
sns.boxplot(data=df, x="PriceCategory", y="SellingPrice")
plt.title("Selling Price by PriceCategory")
plt.show()
```
